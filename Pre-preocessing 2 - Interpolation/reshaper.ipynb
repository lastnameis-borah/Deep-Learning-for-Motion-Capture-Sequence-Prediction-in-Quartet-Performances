# Cello

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the new CSV file into a DataFrame
df_cello_3 = pd.read_csv('Deadpan/Rep1/deadpan_cello_Rep1.csv')


def adjust_data_to_size(data):
    """
    Adjusts a DataFrame to have a shape of 6000x1.
    If the data has more than 6000 rows, it removes the excess.
    If the data has fewer than 6000 rows, it fills with NaN.
    """
    desired_length = 6000
    
    if len(data) > desired_length:
        return data.iloc[:desired_length]
    elif len(data) < desired_length:
        missing_rows = desired_length - len(data)
        nan_df = pd.DataFrame([float('nan')] * missing_rows, columns=data.columns)
        return pd.concat([data, nan_df], ignore_index=True)
    else:
        return data

# Example usage
df = adjust_data_to_size(df_cello_3)
df.shape
```

```python
# Plot the downsampled data from the df_cello_3 DataFrame
plt.figure(figsize=(12, 6))
plt.plot(df['Var1'], label="Downsampled Cello 3 Data", color="purple")
plt.title("Visualization of Downsampled Cello 3 Data")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.to_csv('Deadpan/Rep1/deadpan_cello_Rep1.csv', index=False)
```
