# Deadpan Cello

### Import the modules

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
```

### Load the data

```python
# Reload the original dataset
data_original = pd.read_csv('Deadpan/Rep1/deadpan_cello_Rep1.csv')

data_gb = pd.read_csv('Deadpan/Rep1/deadpan_cello_Rep1.csv')

data_original.shape, data_gb.shape
```

### Apply Gradient Boosting for imputation

```python
# Streamlined Gradient Boosting imputation function
def gb_impute_streamlined(data):
    # Split dataset into train and prediction sets
    train = data.dropna()
    test = data[data.isnull()]
    
    # Use the index as the feature for the model
    X_train = train.index.values.reshape(-1, 1)
    y_train = train.values
    X_test = test.index.values.reshape(-1, 1)
    
    # Train the Gradient Boosting model with different hyperparameters
    model = GradientBoostingRegressor(n_estimators=400, learning_rate=0.05, max_depth=7)
    model.fit(X_train, y_train)
    
    # Predict missing values
    y_pred = model.predict(X_test)
    data.loc[data.isnull()] = y_pred
    
    return data

if data_original['Var1'].isna().any():
    # Apply the streamlined imputation function
    data_gb['Var1'] = gb_impute_streamlined(data_gb['Var1'])

    # Apply moving average for smoothing with a larger window
    data_gb_smoothed_streamlined = data_gb['Var1'].rolling(window=40).mean()
    # Create a combined series: original data where available, otherwise imputed data
    data_combined = data_original['Var1'].combine_first(data_gb['Var1'])
else:
    data_combined = data_original['Var1']
```

### Plot the timeseries data

```python
plt.figure(figsize=(15, 6))
plt.plot(data_combined, label='Imputed Data', color='blue')
plt.plot(data_original['Var1'], alpha=0.5, label='Original Data', color='gray')
plt.title('Time Series Data with Imputed Values')
plt.xlabel('Frame')
plt.ylabel('Position')
plt.legend()
plt.grid(True)
plt.show()
```

### Export into a csv file

```python
data_combined.to_csv('Deadpan/Rep1/deadpan_cello_Rep1_mod.csv', index=False, header=False)
```
